**For GPA**:
 Since GPA is a numeric column, using the mean or median to fill missing values makes sense depending on its distribution.

```python
print(df['GPA'].dtype)
```

```python
df['GPA'] = pd.to_numeric(df['GPA'], errors='coerce')
```

```python
# Recalculate the mean after conversion
mean_gpa = df['GPA'].mean()

# Fill missing values with the mean
df['GPA'].fillna(mean_gpa, inplace=True)

# Print the result
print(f"Filled missing values in GPA with Mean: {mean_gpa}")
```

**For calories_day (Numeric Column):**
calories_day is also numeric, so we handle it similarly by deciding between mean or median.

```python
df['calories_day'] = pd.to_numeric(df['calories_day'], errors='coerce')
```

```python
mean_calories_day = df['calories_day'].mean()
df['calories_day'].fillna(mean_calories_day, inplace=True)
```

```python
print(f"Filled missing values in calories_day with Mean: {mean_calories_day}")
```

**TYPE_SPORTS**:
it is a categorical value.so we cant calculate mean or median for the data.we calculate mode of the data.

```python
df['type_sports'] = df['type_sports'].astype(str)
mode_type_sports = df['type_sports'].mode()[0]
df['type_sports'].fillna(mode_type_sports, inplace=True)
print(f"Filled missing values in type_sports with Mode: {mode_type_sports}")
```

**WEIGHT**:
it is a numeric value.The choice between using the mean or median depends on the distribution of the data.If the data is not heavily skewed, the mean is used because it provides a central value.

```python
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
fill_value = df['weight'].mean() if df['weight'].skew() < 1 else df['weight'].median()
df['weight'].fillna(fill_value, inplace=True)
print(f"Filled missing values in 'weight' with: {fill_value}")
```

**Check for Remaining Missing Values**:
After filling in the missing values, it's good practice to check if there are still any missing values left in the dataset.

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

**Removing Duplicate Rows**:
Now that we've handled missing values, we will check and remove any duplicate rows. This ensures that we have unique records.

```python
initial_row_count = df.shape[0]

```

```python
df.drop_duplicates(inplace=True)
```

```python
new_row_count = df.shape[0]
print(f"\nRemoved {initial_row_count - new_row_count} duplicate rows.")
```

** Removing Duplicate Columns**:
Sometimes, datasets may have duplicate columns (i.e., two columns with the same data).

```python
df = df.loc[:, ~df.columns.duplicated()]
print("\nColumns after removing duplicates:", df.columns.tolist())
```
